```python
import numpy as np
import keras
import AirSimClient
import time
import drone_env

from RL_net import DQNClass
from RL_dataset import Dataset
```

```python
np.set_printoptions(precision=3, suppress=True)
test_model = None
test_model = keras.models.load_model("test.h5")
```

```python
settings = {}

settings["learning_rate"] = 0.01
settings["reward_decay"] = 0.9
settings["e_greedy"] = 0.9
settings["e_greedy_increasement"] = 0.01
settings["memory_length"] = 8192
settings["batch_size"] = 32
settings["epochs"] = 1
settings["replace_target_iter"] = 32
settings["model"] = test_model
settings["n_actions"] = 7
settings["n_features"] = 6

aim = AirSimClient.Vector3r(32,38,-4)
start = AirSimClient.Vector3r(0,0,-5)
```

```python
RL = DQNClass(settings)
env = drone_env.drone_env_gridworld()
```

```python


current = 0
count = 0
learns = 0
epochs = 0
succ = 0
learn = 0


s = env.reset()
while True:
    
    action = RL.choose_action(s)
    
    s_,r,done,info = env.step(action)
    
    env.render(extra1 = str(count),extra2 = "reward: "+str(int(r))+" Done: "+str(succ))
      
    count += 1
    current += 1
    if info == "success":
        succ += 1
    if current>100:
        done = True
        info = "out of steps"
    
    counter = RL.add_data(s, action, r, s_)
    
    if done:
        current = 0
        epochs += 1
        RL.model.save("test.h5")
        env.reset()
        print ("reset due to",info)
        print ("--------------------------------reset---------------------------")
        
    if count > 10:
        RL.learn(times = 10)
        learn += 1
        
    s = s_
```
